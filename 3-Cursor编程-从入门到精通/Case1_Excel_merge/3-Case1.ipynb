# Case 1 - Excel merge

```python
# Reading Basic Excel Form
basic_info_df = pd.read_excel("员工基本信息表.xlsx")
# Reading Performance Excel Form
performance_df = pd.read_excel("员工绩效表.xlsx")
```

```python
# Check the basic info
print(basic_info_df.head())
print("\n")
print(basic_info_df.columns)
print("\n\n")
print(performance_df.head())
print("\n")
print(performance_df.columns)

```

```python
# select the data from 2024 Q4
q4_2024_perf = performance_df[(performance_df['年度']== 2024) & (performance_df['季度'] == 4)]
#check the data
print(q4_2024_perf.head())
```

```python
# Merge the data
merged_df = pd.merge(basic_info_df, 
                    q4_2024_perf[['员工ID', '绩效评分']], 
                    on='员工ID', 
                    how='left')

# rename the column
merged_df = merged_df.rename(columns={'绩效评分': '2024年第四季度绩效评分'})


#check the data
print(merged_df.head())
print("\n")
print(merged_df.columns)
```
